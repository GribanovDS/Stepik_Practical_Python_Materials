### 2. Загрузка датасета


MNE-Python предоставляет встроенные примерные данные.
Мы будем использовать `sample`-датасет, который содержит MEG/EEG запись аудиовизуального эксперимента – https://mne.tools/stable/auto_tutorials/intro/10_overview.html

При первом запуске данные будут автоматически скачаны (по умолчанию в `~/mne_data`).

```python
# Путь к sample-датасету (если данных нет, они автоматически скачаются)
sample_data_folder = mne.datasets.sample.data_path()
raw_path = os.path.join(
    sample_data_folder,
    "MEG",
    "sample",
    "sample_audvis_raw.fif",
)
```

Проверим, что данные скачались. Должен напечататься путь до файла с данными

```python
raw_path
```

Примеры форматов ЭЭГ данных

- `.fif` (Functional Imaging File) — «родной» формат MNE, умеет хранить не только сырые сигналы, но и полную мета‑информацию об эксперименте, аннотации и результаты обработки и тд.

- `.edf` (European Data Format) — европейский стандарт: текстовый заголовок + бинарные данные каналов. Подходит как «универсальный» формат обмена между разными устройствами и программами, но хранит меньше структурированной информации.

- `.bdf` (Biosemi Data Format) — очень похож на EDF, но с 24‑битным представлением отсчётов для большего динамического диапазона; используется в основном с усилителями BioSemi.

### 3. Загрузка Raw-объекта


Теперь загрузим сырые данные в объект `Raw`, используя функции из [`mne.io`](https://mne.tools/stable/api/reading_raw_data.html#module-mne.io)
Мы сразу обрежем запись до первых 60 секунд и загрузим её в память (`preload=True`) для ускорения дальнейших операций.

```python
raw = mne.io.read_raw_fif(raw_path, preload=True).crop(tmax=60)
```

```python
print(raw)
```

Отобразим с помощью аттрибута `raw.info` содержит метаданные: список каналов, частота дискретизации, типы каналов (EEG/MEG/EOG/STIM и т.п.), фильтры, аннотации и др.


```python
raw.info
```

В Python у объекта есть:

- Атрибуты (свойства) — «прикреплённые переменные», в них лежат данные. У `Raw` таким атрибутом является `info:` словарь/структура с мета‑информацией

- Методы — «прикреплённые функции», которые что‑то делают с объектом, например `mne.io.read_raw_fif(...)`

Поэтому:

- `raw.info` — просто обращение к объекту, скобок нет.

- `mne.io.read_raw_fif(...)` — вызов метода‑функции, поэтому нужны скобки.

## 4. Базовая визуализация сырых данных

Теперь посмотрим на временные ряды каналов с помощью `raw.plot()`

```python
raw.plot(n_channels=2, duration=5.0, picks='eeg', scalings=1/10000)
```

`raw.plot(...)` внутри себя уже показывает фигуру и при этом одновременно возвращает объект фигуры – поэтому имеем два графика

```python
_ = raw.plot(n_channels=2, duration=5.0, picks='eeg', scalings=1/10000)
```

Вместо `1/10000` можно использовать `1e-4`, где буква «e» в научной нотации заменяет «умножить на 10 в степени»

```python
_ = raw.plot(n_channels=2, duration=5.0, picks='eeg', scalings=1e-4)
```

### 2.3 Сколько длится наш сигнал?

Знакомимся с атрибутами `mne.Raw.info` по ссылке: https://mne.tools/stable/generated/mne.Info.html#mne.Info

**Частота дискретизации** (sampling frequency) – это количество измерений (сэмплов) аналогового сигнала, производимых за одну секунду при его преобразовании в цифровой формат. Измеряется в Герцах (Гц)

Вычислите частоту дискретизации у данных с помощью аттрибута `info['sfreq']`

А теперь разделите `raw.n_times` на частоту дискретизации и округлите с помозью функции `round(...)`

```python
round(raw.n_times / raw.info['sfreq'])
```

Теперь придумайте объяснение, почему в пунктах 2.1, 2.2 и 2.3 получаются одинаковые результаты
